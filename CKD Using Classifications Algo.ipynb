```python
#importing the Libraies
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Reading the Dataset
dataset = pd.read_csv('CKD.csv')
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset,drop_first=True,dtype = int)
```

```python
dataset
```

```python
dataset.columns
```

```python
indep=dataset[['age', 'bp', 'al', 'su', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hrmo', 'pcv',
       'wc', 'rc', 'sg_b', 'sg_c', 'sg_d', 'sg_e', 'rbc_normal', 'pc_normal',
       'pcc_present', 'ba_present', 'htn_yes', 'dm_yes', 'cad_yes',
       'appet_yes', 'pe_yes', 'ane_yes']]
dep=dataset['classification_yes']
```

```python
#split into training set and test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(indep, dep, test_size = 1/3, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
#Create a Class
models_results = {}
#Create a Function
def evaluate_model(model, params, model_name):
    print(f"\nüìå Evaluating {model_name}...\n")
    #Using If_else Condition to create a model
    if params:
        grid = GridSearchCV(model, params, cv=5, scoring='accuracy', n_jobs=-1)
        grid.fit(X_train, y_train)
        best_model = grid.best_estimator_
        print("Best Parameters:", grid.best_params_)
    else:
        best_model = model
        best_model.fit(X_train, y_train)
    
    y_pred = best_model.predict(X_test)
    acc = accuracy_score(y_test, y_pred)
    print("Accuracy:", acc)
    print("Classification Report:\n", classification_report(y_test, y_pred))
    print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
    
    # Store results
    models_results[model_name] = {
        'model': best_model,
        'accuracy': acc
    }

# 1. Logistic Regression
from sklearn.linear_model import LogisticRegression
evaluate_model(LogisticRegression(max_iter=1000), {
    'C': [0.01, 0.1, 1, 10],
    'solver': ['liblinear', 'lbfgs']
}, "Logistic Regression")

# 2. K-Nearest Neighbors
from sklearn.neighbors import KNeighborsClassifier
evaluate_model(KNeighborsClassifier(), {
    'n_neighbors': [3, 5, 7, 9],
    'weights': ['uniform', 'distance'],
    'metric': ['euclidean', 'manhattan']
}, "K-Nearest Neighbors")

# 3. Decision Tree
from sklearn.tree import DecisionTreeClassifier
evaluate_model(DecisionTreeClassifier(), {
    'criterion': ['gini', 'entropy'],
    'max_depth': [None, 5, 10, 20],
    'min_samples_split': [2, 5, 10]
}, "Decision Tree")

# 4. Random Forest
from sklearn.ensemble import RandomForestClassifier
evaluate_model(RandomForestClassifier(), {
    'n_estimators': [50, 100, 200],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5],
    'criterion': ['gini', 'entropy']
}, "Random Forest")

# 5. Support Vector Machine
from sklearn.svm import SVC
evaluate_model(SVC(), {
    'C': [0.1, 1, 10],
    'kernel': ['linear', 'rbf', 'poly'],
    'gamma': ['scale', 'auto']
}, "Support Vector Machine")

# 6. Naive Bayes (no hyperparameters needed for GaussianNB)
from sklearn.naive_bayes import GaussianNB
evaluate_model(GaussianNB(), None, "Naive Bayes")
```

```python
# Convert results into a DataFrame for easy comparison
results_df = pd.DataFrame([
    {'Model': model, 
     'Accuracy': round(models_results[model]['accuracy'], 4),
     'Best Parameters': str(models_results[model]['model'].get_params())}
    for model in models_results
])

# Sort by accuracy
results_df = results_df.sort_values(by='Accuracy', ascending=False).reset_index(drop=True)

# Show the comparison table
print("üìä Model Comparison Summary:")
print(results_df)
```

```python
from sklearn.metrics import roc_auc_score

# Get the best model from your stored results (for example, using Random Forest)
best_model = models_results["Random Forest"]['model']

# Calculate ROC AUC score
roc_auc = roc_auc_score(y_test, best_model.predict_proba(X_test)[:,1])
print(f"ROC AUC Score: {roc_auc}")
```

```python
import pandas as pd
import numpy as np

def predict_ckd():
    # Create a dictionary to store user inputs
    input_data = {}
    
    # Numerical features
    print("Please enter the following numerical values:")
    input_data['age'] = float(input("Age: "))
    input_data['bp'] = float(input("Blood Pressure (bp): "))
    input_data['al'] = float(input("Albumin (al): "))
    input_data['su'] = float(input("Sugar (su): "))
    input_data['bgr'] = float(input("Blood Glucose Random (bgr): "))
    input_data['bu'] = float(input("Blood Urea (bu): "))
    input_data['sc'] = float(input("Serum Creatinine (sc): "))
    input_data['sod'] = float(input("Sodium (sod): "))
    input_data['pot'] = float(input("Potassium (pot): "))
    input_data['hrmo'] = float(input("Hemoglobin (hrmo): "))
    input_data['pcv'] = float(input("Packed Cell Volume (pcv): "))
    input_data['wc'] = float(input("White Blood Cell Count (wc): "))
    input_data['rc'] = float(input("Red Blood Cell Count (rc): "))
    
    # Categorical features
    print("\nPlease answer the following categorical questions (yes/no):")
    input_data['sg_b'] = 1 if input("Specific Gravity - b (yes/no): ").lower() == 'yes' else 0
    input_data['sg_c'] = 1 if input("Specific Gravity - c (yes/no): ").lower() == 'yes' else 0
    input_data['sg_d'] = 1 if input("Specific Gravity - d (yes/no): ").lower() == 'yes' else 0
    input_data['sg_e'] = 1 if input("Specific Gravity - e (yes/no): ").lower() == 'yes' else 0
    input_data['rbc_normal'] = 1 if input("Red Blood Cells normal (yes/no): ").lower() == 'yes' else 0
    input_data['pc_normal'] = 1 if input("Pus Cell normal (yes/no): ").lower() == 'yes' else 0
    input_data['pcc_present'] = 1 if input("Pus Cell Clumps present (yes/no): ").lower() == 'yes' else 0
    input_data['ba_present'] = 1 if input("Bacteria present (yes/no): ").lower() == 'yes' else 0
    input_data['htn_yes'] = 1 if input("Hypertension (yes/no): ").lower() == 'yes' else 0
    input_data['dm_yes'] = 1 if input("Diabetes Mellitus (yes/no): ").lower() == 'yes' else 0
    input_data['cad_yes'] = 1 if input("Coronary Artery Disease (yes/no): ").lower() == 'yes' else 0
    input_data['appet_yes'] = 1 if input("Good Appetite (yes/no): ").lower() == 'yes' else 0
    input_data['pe_yes'] = 1 if input("Pedal Edema (yes/no): ").lower() == 'yes' else 0
    input_data['ane_yes'] = 1 if input("Anemia (yes/no): ").lower() == 'yes' else 0
    
    # Convert to DataFrame
    input_df = pd.DataFrame([input_data])
    
    # Ensure correct column order (same as training data)
    columns_order = ['age', 'bp', 'al', 'su', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hrmo', 'pcv',
                   'wc', 'rc', 'sg_b', 'sg_c', 'sg_d', 'sg_e', 'rbc_normal', 'pc_normal',
                   'pcc_present', 'ba_present', 'htn_yes', 'dm_yes', 'cad_yes',
                   'appet_yes', 'pe_yes', 'ane_yes']
    input_df = input_df[columns_order]
    
    # Scale the input data (using the same scaler from training)
    input_scaled = sc.transform(input_df)
    
    # Make prediction (using your best model)
    prediction = models_results["Random Forest"]['model'].predict(input_scaled)
    probability = models_results["Random Forest"]['model'].predict_proba(input_scaled)
    
    # Display results
    print("\nPrediction Results:")
    print("-------------------")
    print(f"Predicted Classification: {'CKD (yes)' if prediction[0] == 1 else 'No CKD (no)'}")
    print(f"Probability of CKD: {probability[0][1]*100:.2f}%")
    print(f"Probability of No CKD: {probability[0][0]*100:.2f}%")

# Call the function to make a prediction
predict_ckd()
```
